**LAB2: Simulation of the R-C amplifier circuit (Continuous System)**

**Objectives:
To develop a mathematical model of an RC Amplifier Circuit**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function1(a11,v10,ein):
  return a11*v10-a11*ein        # since a11=-b11, we use -a11
```

```python
def function2(a21,a22,v10,v20,ein):
  return a21*v10+a22*v20-a21*ein   #since a21=-b21, we use -a21
```

```python
#initial conditions

v10=0
v20=0
h=0.0002
ein=1.5
a11=-50
a21=-10000
a22=-21.5

v1=np.zeros(500)
v2=np.zeros(500)
t=np.zeros(500)
```

```python
for i in range(500):
    m11=function1(a11, v10, ein);
    m12=function1(a11,v10+m11*h/2,ein);
    m13=function1(a11,v10+m12*h/2,ein);
    m14=function1(a11,v10+m13*h,ein);
    v11=v10+((m11+2*m12+2*m13+m14)/6)*h;

    m21=function2(a21, a22, v10,v20,ein);
    m22=function2(a21, a22, v10+h/2,v20+m21*h/2,ein);
    m23=function2(a21, a22, v10+h/2,v20+m22*h/2,ein);
    m24=function2(a21, a22, v10+h,v20+m23*h,ein);
    v21=v20+((m21+2*m22+2*m23+m24)/6)*h;
    v10=v11;
    v20=v21;
    t[i] == h*i;
    v1[i] == v10;
    v2[i] == v20;

plt.figure(figsize=(10, 6))
plt.plot(t, v1, label='v1')
plt.title('v1 over Time')
plt.xlabel('Time')
plt.ylabel('v1')
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(t, v2, label='v2')
plt.title('v2 over Time')
plt.xlabel('Time')
plt.ylabel('v2')
plt.legend()
plt.show()
```
