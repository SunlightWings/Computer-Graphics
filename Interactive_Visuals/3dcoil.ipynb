```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np

# Function to create a 3D surface
def flashy_surface_animation(frames=100):
    t = np.linspace(0, 4 * np.pi, frames)
    x = 5 * np.sin(t)
    y = 5 * np.cos(t)
    z = t

    fig = go.Figure()

    for i in range(frames):
        fig.add_trace(go.Scatter3d(
            x=x,
            y=y,
            z=z[:i + 1],
            mode='markers',
            marker=dict(size=8, color=z[:i + 1], colorscale='Viridis', opacity=0.8),
            name=f'Frame {i + 1}'
        ))

    fig.update_layout(scene=dict(
        xaxis=dict(title='X Axis'),
        yaxis=dict(title='Y Axis'),
        zaxis=dict(title='Z Axis'),
    ))

    return fig

# Display the surface animation
animation_fig = flashy_surface_animation()
animation_fig.show()
```
