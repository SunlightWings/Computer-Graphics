```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Generate all possible combinations of red, green, and blue values
red_values = np.arange(0, 10)
green_values = np.arange(0, 10)
blue_values = np.arange(0, 10)

# Create a meshgrid for all combinations
red, green, blue = np.meshgrid(red_values, green_values, blue_values)

# Flatten the arrays for plotting
red_flat = red.flatten()
green_flat = green.flatten()
blue_flat = blue.flatten()

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(red_flat, green_flat, blue_flat, c=np.vstack((red_flat, green_flat, blue_flat)).T / 10.0)

# Set axis labels
ax.set_xlabel('Red')
ax.set_ylabel('Green')
ax.set_zlabel('Blue')

plt.title('RGB Color Space')
plt.show()
```
