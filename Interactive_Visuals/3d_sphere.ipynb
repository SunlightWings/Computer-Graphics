```python
# Create sample data
theta = np.linspace(0, 2*np.pi, 100)
phi = np.linspace(0, np.pi, 50)
theta, phi = np.meshgrid(theta, phi)
```

```python
# Define the function to map to the sphere surface
r = 1 + 0.5 * np.sin(phi) * np.cos(theta)
x = r * np.sin(phi) * np.cos(theta)
y = r * np.sin(phi) * np.sin(theta)
z = r * np.cos(phi)

# Create a color variable based on the surface height
color_variable = x**2 + y**2 + z**2
```

```python
# Create 3D scatter plot on the sphere surface
fig = go.Figure(data=[go.Scatter3d(x=x.flatten(), y=y.flatten(), z=z.flatten(), mode='markers',
                                   marker=dict(size=5, color=color_variable, colorscale='Viridis'))])

# Customize the layout
fig.update_layout(title='3D Sphere with Variable Colors',
                  scene=dict(xaxis_title='X Axis', yaxis_title='Y Axis', zaxis_title='Z Axis'))

# Show the plot
fig.show()
```
